# 3. Feature Engineering

```python
# max, avg bids grouped by auction, country, device, ip , url

max_avg_agg = pd.DataFrame()

max_avg_agg['max_bids_auc']=bids_df.groupby(['bidder_id', 'auction'])['bid_id'].count().max(level=0)
max_avg_agg[ 'avg_bids_auc']=bids_df.groupby(['bidder_id', 'auction'])['bid_id'].count().mean(level=0)

max_avg_agg['max_bids_dev']=bids_df.groupby(['bidder_id', 'device'])['bid_id'].count().max(level=0)
max_avg_agg[ 'avg_bids_dev']=bids_df.groupby(['bidder_id', 'device'])['bid_id'].count().mean(level=0)

max_avg_agg['max_bids_country']=bids_df.groupby(['bidder_id', 'country'])['bid_id'].count().max(level=0)
max_avg_agg[ 'avg_bids_country']=bids_df.groupby(['bidder_id', 'country'])['bid_id'].count().mean(level=0)

max_avg_agg['max_bids_ip']=bids_df.groupby(['bidder_id', 'ip'])['bid_id'].count().max(level=0)
max_avg_agg[ 'avg_bids_ip']=bids_df.groupby(['bidder_id', 'ip'])['bid_id'].count().mean(level=0)

max_avg_agg['max_bids_url']=bids_df.groupby(['bidder_id', 'url'])['bid_id'].count().max(level=0)
max_avg_agg[ 'avg_bids_url']=bids_df.groupby(['bidder_id', 'url'])['bid_id'].count().mean(level=0)

max_avg_agg = max_avg_agg.reset_index()
max_avg_agg
```

```python
##sort the bids in ascending order of bidder_ids and time.

bids_sorted_df = bids_df.sort_values(['bidder_id', 'time'], ascending = [True, True])
bids_sorted_df.head()
```

```python
## compute the time differences between bids

timediff = bids_sorted_df.groupby('bidder_id')['time'].diff()
timediff= timediff.fillna('0')

bids_sorted_df['timediff'] = timediff
bids_sorted_df['timediff'] = bids_sorted_df['timediff'].astype(float)

bids_sorted_df.head()
```

```python
time_agg = pd.DataFrame()

time_agg[['maxtime', 'mintime']] = bids_sorted_df.groupby('bidder_id')['time'].agg(['max','min'])
time_agg[['maxtimediff', 'meantimediff', 'mediantimediff']] = bids_sorted_df.groupby('bidder_id')['timediff'].agg(['max','mean','median'])
time_agg['duration']=time_agg['maxtime']-time_agg['mintime']
time_agg = time_agg.reset_index()
time_agg
```

```python
##Group features by no. of unique bidder ids
bidder_unique = bids_df.groupby("bidder_id").nunique()
bidder_unique
```

```python
bidder_unique = bidder_unique.drop(columns='bidder_id', errors='ignore').reset_index()
bidder_unique
```

```python
##Merge unique bidder id table with max_avg table
bids_agg = bidder_unique.merge(max_avg_agg, on='bidder_id', how='left')
bids_agg
```

```python
##Merge aggregated bids table with time difference table
bids_aggr = bids_agg.merge(time_agg, on='bidder_id', how='left')

bids_aggr['bids_per_auction']= bids_aggr['bid_id']/bids_aggr['auction']
bids_aggr['bids_per_device']= bids_aggr['bid_id']/bids_aggr['device']
bids_aggr['bids_per_country']= bids_aggr['bid_id']/bids_aggr['country']
bids_aggr['bids_per_ip']= bids_aggr['bid_id']/bids_aggr['ip']
bids_aggr['bids_per_url']= bids_aggr['bid_id']/bids_aggr['url']
bids_aggr
```

```python
##Combine table of all aggregated features with train and test dataset
train_set = train_df.merge(bids_aggr, on='bidder_id', how='left')
test_set = test_df.merge(bids_aggr, on='bidder_id', how='left')
```

```python
##Replace null values with 0s
train_set = train_set.fillna(0)
test_set = test_set.fillna(0)
```

# 4. Model Training

```python
##Selecting feature columns for model validation
feature_cols = ['bid_id','auction','merchandise','device','time','country','ip','url','max_bids_auc', 'avg_bids_auc',
                'max_bids_dev','avg_bids_dev','max_bids_country','avg_bids_country','max_bids_ip','avg_bids_ip',
                'max_bids_url','avg_bids_url','maxtime','mintime','maxtimediff','meantimediff','mediantimediff','duration',
     'bids_per_auction']

X = train_set[feature_cols]
y = train_set['outcome']
```

```python
##Importing selected models and libraries
from imblearn.pipeline import Pipeline, make_pipeline
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn import metrics
import statistics
```

```python
## manual oversampling in CV
```

```python
#Model validation
from catboost import CatBoostClassifier
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')
#Ensemble model of RandomForest, Xgboost and catboost
rf_model = RandomForestClassifier(n_estimators=100, max_depth=4,random_state=0, max_features="log2")
cat_model = CatBoostClassifier(random_seed = 0, depth = 4,verbose=False) #default depth =6
xgb_model = xgb.XGBClassifier(random_state=0, max_depth=4)
    
kf = KFold(n_splits=10)
smoter = SMOTE(random_state=0)
    
k_fold_auc = []

    # Iterate through each fold and calculate the auc, accuracy for each fold
for train_index, test_index in kf.split(X, y):
        
        # Extract the training and test data
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Upsample only the data in the training section
        X_train_upsample, y_train_upsample = smoter.fit_resample(X_train,y_train)
        
        # Fit the model on the upsampled training data
        rf_model.fit(X_train_upsample, y_train_upsample)
        cat_model.fit(X_train_upsample, y_train_upsample)
        xgb_model.fit(X_train_upsample, y_train_upsample)
        
        y_pred1 = rf_model.predict(X_test)
        y_pred2 = cat_model.predict(X_test)
        y_pred3 = xgb_model.predict(X_test)
        
        #Calculate Accuracy, AUC for the fold and append it
        
        final_pred = (y_pred1 * 0.6) + (y_pred2 * 0.3) + (y_pred3 * 0.1)

        auc = metrics.roc_auc_score(y_test, final_pred)

        k_fold_auc.append(auc)

print('The average auc is:', statistics.mean(k_fold_auc))
```
